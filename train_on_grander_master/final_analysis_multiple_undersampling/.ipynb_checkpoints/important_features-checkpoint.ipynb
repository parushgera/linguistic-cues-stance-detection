# Strict Filter

1. Statistically significantly different in ≥7 iterations AND
2. Same LR coefficient sign across above ≥7 iterations AND
3. Exist in BOTH LR and XGBoost with non-zero values

```python
import pandas as pd
import numpy as np
import os
import warnings
warnings.filterwarnings('ignore')

def parse_mean_std(mean_std_str):
    """Parse '0.1234 ± 0.0567' format"""
    if pd.isna(mean_std_str) or mean_std_str == '':
        return np.nan, np.nan
    parts = str(mean_std_str).split(' ± ')
    if len(parts) == 2:
        return float(parts[0]), float(parts[1])
    return np.nan, np.nan

def create_final_aggregated_features_strict(base_dir, output_dir, 
                                          min_significance_count=7,
                                          total_iterations=13):
    """
    Aggregate features with STRICT filtering - ALL criteria must be met:
    1. Statistically significantly different in ≥MIN_SIGNIFICANCE_COUNT iterations
    2. Same LR coefficient sign across iterations
    3. Exist in BOTH LR and XGBoost with non-zero values
    4. Non-zero values for both models
    
    Parameters:
    -----------
    base_dir : str
        Base directory containing iteration folders
    output_dir : str  
        Output directory for final files
    min_significance_count : int
        Minimum number of iterations for significance (default: 7)
    total_iterations : int
        Total number of iterations (default: 13)
    """
    stances = ["All_Stances", "FAVOR", "AGAINST", "NONE"]
    models = ["Logistic Regression", "XGBoost"]
    
    os.makedirs(output_dir, exist_ok=True)
    
    print(f"Configuration - STRICT FILTERING (ALL criteria must be met):")
    print(f"  1. Statistically significantly different in ≥{min_significance_count} iterations")
    print(f"  2. Same LR coefficient sign across iterations")
    print(f"  3. Exist in BOTH LR and XGBoost with non-zero values")
    print(f"  4. Non-zero values for both models")
    print(f"  Total iterations: {total_iterations}")
    
    summary_stats = []
    
    for stance in stances:
        print(f"\nProcessing {stance}...")
        
        # Step 1: Collect all features from all iterations
        all_feature_data = {}  # feature_name -> list of data from each iteration
        
        for i in range(1, total_iterations + 1):
            file_path = os.path.join(base_dir, f"{i:02d}_iteration", f"top_features_{stance}.csv")
            if os.path.exists(file_path):
                df = pd.read_csv(file_path)
                print(f"  Loaded iteration {i:02d}: {len(df)} features")
                
                for _, row in df.iterrows():
                    # Get clean feature name (without *)
                    clean_feature = str(row['feature']).replace('*', '')
                    
                    if clean_feature not in all_feature_data:
                        all_feature_data[clean_feature] = []
                    
                    # Store all data for this feature occurrence
                    feature_occurrence = {
                        'iteration': i,
                        'is_significant': '*' in str(row['feature']),
                        'coefficients': {},
                        'correct_mean_std': row.get('correct_mean_std', ''),
                        'incorrect_mean_std': row.get('incorrect_mean_std', '')
                    }
                    
                    # Store coefficients/importances (only if they are non-zero)
                    for model in models:
                        if (model in df.columns and 
                            pd.notna(row[model]) and 
                            row[model] != '' and 
                            row[model] != 0):
                            feature_occurrence['coefficients'][model] = float(row[model])
                    
                    all_feature_data[clean_feature].append(feature_occurrence)
            else:
                print(f"  Warning: Missing file {file_path}")
        
        print(f"  Total features found: {len(all_feature_data)}")
        
        # Step 2: Apply STRICT filtering - ALL criteria must be met
        final_features = []
        
        for feature_name, occurrences in all_feature_data.items():
            # Aggregate data across occurrences
            aggregated_data = {
                'feature_name': feature_name,
                'total_occurrences': len(occurrences),
                'coefficients': {model: [] for model in models},
                'coef_occurrences': {model: 0 for model in models},
                'significances': [],
                'correct_means': [],
                'incorrect_means': [],
                'correct_stds': [],
                'incorrect_stds': []
            }
            
            # Collect data from all occurrences
            for occurrence in occurrences:
                aggregated_data['significances'].append(occurrence['is_significant'])
                
                # Collect coefficients and count valid occurrences
                for model in models:
                    if model in occurrence['coefficients']:
                        aggregated_data['coefficients'][model].append(occurrence['coefficients'][model])
                        aggregated_data['coef_occurrences'][model] += 1
                
                # Parse mean/std values
                if occurrence['correct_mean_std']:
                    correct_mean, correct_std = parse_mean_std(occurrence['correct_mean_std'])
                    if not np.isnan(correct_mean):
                        aggregated_data['correct_means'].append(correct_mean)
                        aggregated_data['correct_stds'].append(correct_std)
                
                if occurrence['incorrect_mean_std']:
                    incorrect_mean, incorrect_std = parse_mean_std(occurrence['incorrect_mean_std'])
                    if not np.isnan(incorrect_mean):
                        aggregated_data['incorrect_means'].append(incorrect_mean)
                        aggregated_data['incorrect_stds'].append(incorrect_std)
            
            # CRITERION 1: Check significance threshold
            num_significant = sum(aggregated_data['significances'])
            if num_significant < min_significance_count:
                print(f"    Skipping {feature_name}: only {num_significant}/{len(occurrences)} significant (need ≥{min_significance_count})")
                continue
            
            # CRITERION 3 & 4: Must have non-zero values in BOTH models
            lr_coeffs = aggregated_data['coefficients']['Logistic Regression']
            xgb_coeffs = aggregated_data['coefficients']['XGBoost']
            
            if not lr_coeffs:
                print(f"    Skipping {feature_name}: no LR coefficients")
                continue
            
            if not xgb_coeffs:
                print(f"    Skipping {feature_name}: no XGBoost importance values")
                continue
            
            # CRITERION 2: Check LR coefficient sign consistency
            lr_signs = [1 if x > 0 else -1 if x < 0 else 0 for x in lr_coeffs]
            lr_non_zero_signs = [s for s in lr_signs if s != 0]
            
            if len(set(lr_non_zero_signs)) > 1:
                print(f"    Skipping {feature_name}: LR mixed signs ({len(lr_coeffs)} values)")
                continue
            
            if len(lr_non_zero_signs) == 0:
                print(f"    Skipping {feature_name}: all LR coefficients are zero")
                continue
            
            # ALL CRITERIA MET - Calculate final values
            final_lr_coef = np.median(lr_coeffs)
            final_xgb_importance = np.median(xgb_coeffs)
            
            # Calculate aggregated statistics
            agg_correct_mean = np.mean(aggregated_data['correct_means']) if aggregated_data['correct_means'] else np.nan
            agg_correct_std = np.mean(aggregated_data['correct_stds']) if aggregated_data['correct_stds'] else np.nan
            agg_incorrect_mean = np.mean(aggregated_data['incorrect_means']) if aggregated_data['incorrect_means'] else np.nan
            agg_incorrect_std = np.mean(aggregated_data['incorrect_stds']) if aggregated_data['incorrect_stds'] else np.nan
            
            # Create final entry
            final_features.append({
                'feature': feature_name + '*',  # All included features meet significance threshold
                'Logistic Regression': final_lr_coef,
                'XGBoost': final_xgb_importance,
                'correct_mean_std': f"{agg_correct_mean:.4f} ± {agg_correct_std:.4f}" if not np.isnan(agg_correct_mean) else '',
                'incorrect_mean_std': f"{agg_incorrect_mean:.4f} ± {agg_incorrect_std:.4f}" if not np.isnan(agg_incorrect_mean) else '',
                'lr_coef_count': len(lr_coeffs),
                'xgb_coef_count': len(xgb_coeffs),
                'num_significant': num_significant,
                'significance_rate': f"{num_significant}/{len(occurrences)}",
                'lr_sign': 'positive' if lr_non_zero_signs[0] > 0 else 'negative'
            })
        
        # Step 3: Sort by significance count, then by coefficient counts
        final_features.sort(key=lambda x: (x['num_significant'], x['lr_coef_count'] + x['xgb_coef_count']), reverse=True)
        
        # Remove helper columns for final output
        final_features_clean = []
        for feature in final_features:
            clean_feature = {k: v for k, v in feature.items() 
                           if k not in ['lr_coef_count', 'xgb_coef_count', 'num_significant', 'significance_rate', 'lr_sign']}
            final_features_clean.append(clean_feature)
        
        # Step 4: Save final aggregated file
        final_df = pd.DataFrame(final_features_clean)
        output_file = os.path.join(output_dir, f"final_top_features_{stance}.csv")
        final_df.to_csv(output_file, index=False)
        
        # Step 5: Save detailed info file with occurrence counts
        detailed_df = pd.DataFrame(final_features)
        detailed_file = os.path.join(output_dir, f"detailed_features_{stance}.csv")
        detailed_df.to_csv(detailed_file, index=False)
        
        # Collect summary statistics
        total_features = len(all_feature_data)
        final_count = len(final_features)
        positive_lr = len([f for f in final_features if f['lr_sign'] == 'positive'])
        negative_lr = len([f for f in final_features if f['lr_sign'] == 'negative'])
        
        summary_stats.append({
            'Stance': stance,
            'Total_Features': total_features,
            'Final_Features': final_count,
            'Positive_LR_Coef': positive_lr,
            'Negative_LR_Coef': negative_lr,
            'Avg_Significance': round(np.mean([f['num_significant'] for f in final_features]) if final_features else 0, 1),
            'Success_Rate': f"{final_count}/{total_features}",
            'Success_Percentage': f"{100*final_count/total_features:.1f}%" if total_features > 0 else "0%"
        })
        
        print(f"  Final features meeting ALL criteria: {final_count}")
        print(f"  Positive LR coefficients: {positive_lr}")
        print(f"  Negative LR coefficients: {negative_lr}")
        print(f"  Average significance: {round(np.mean([f['num_significant'] for f in final_features]) if final_features else 0, 1)}/{total_iterations}")
        print(f"  Success rate: {100*final_count/total_features:.1f}%")
        print(f"  Saved to: {output_file}")
        print(f"  Detailed info saved to: {detailed_file}")
    
    # Save summary statistics
    summary_df = pd.DataFrame(summary_stats)
    summary_file = os.path.join(output_dir, f"aggregation_summary_strict_{min_significance_count}sig.csv")
    summary_df.to_csv(summary_file, index=False)
    
    print(f"\n" + "="*80)
    print(f"AGGREGATION SUMMARY - STRICT FILTERING (ALL 4 criteria)")
    print("="*80)
    print(summary_df.to_string(index=False))
    print(f"\nSummary saved to: {summary_file}")
    
    return summary_df

# Main execution with configurable parameters
if __name__ == '__main__':
    base_dir = ""
    output_dir = "important_features"
    
    # CONFIGURABLE PARAMETERS - CHANGE THESE AS NEEDED
    MIN_SIGNIFICANCE_COUNT = 7   # Feature must be significant in at least 7 iterations  
    TOTAL_ITERATIONS = 13        # Total number of iterations
    
    print("Creating final aggregated feature files with STRICT filtering...")
    print("ALL 4 criteria must be met:")
    print("1. Statistically significantly different in ≥7 iterations")
    print("2. Same LR coefficient sign across iterations") 
    print("3. Exist in BOTH LR and XGBoost with non-zero values")
    print("4. Non-zero values for both models")
    
    summary = create_final_aggregated_features_strict(
        base_dir=base_dir,
        output_dir=output_dir,
        min_significance_count=MIN_SIGNIFICANCE_COUNT,
        total_iterations=TOTAL_ITERATIONS
    )
    
    print(f"\n✅ Strict filtering completed!")
    print(f"📁 Created files:")
    print(f"   - final_top_features_*.csv (4 files) - only features meeting ALL criteria")
    print(f"   - detailed_features_*.csv (4 files with detailed diagnostics)")
    print(f"   - aggregation_summary_strict_{MIN_SIGNIFICANCE_COUNT}sig.csv")
```

# Only LR

```python
import pandas as pd
import numpy as np
import os
import warnings
warnings.filterwarnings('ignore')

def parse_mean_std(mean_std_str):
    """Parse '0.1234 ± 0.0567' format"""
    if pd.isna(mean_std_str) or mean_std_str == '':
        return np.nan, np.nan
    parts = str(mean_std_str).split(' ± ')
    if len(parts) == 2:
        return float(parts[0]), float(parts[1])
    return np.nan, np.nan

def create_final_aggregated_features_lr_only(base_dir, output_dir, 
                                           min_significance_count=7,
                                           total_iterations=13):
    """
    Aggregate features with LR-ONLY filtering:
    1. Statistically significantly different in ≥MIN_SIGNIFICANCE_COUNT iterations
    2. Same LR coefficient sign across iterations
    3. Non-zero LR coefficients (XGBoost requirements ignored)
    
    Parameters:
    -----------
    base_dir : str
        Base directory containing iteration folders
    output_dir : str  
        Output directory for final files
    min_significance_count : int
        Minimum number of iterations for significance (default: 7)
    total_iterations : int
        Total number of iterations (default: 13)
    """
    stances = ["All_Stances", "FAVOR", "AGAINST", "NONE"]
    models = ["Logistic Regression", "XGBoost"]
    
    os.makedirs(output_dir, exist_ok=True)
    
    print(f"Configuration - LR-ONLY FILTERING:")
    print(f"  1. Statistically significantly different in ≥{min_significance_count} iterations")
    print(f"  2. Same LR coefficient sign across iterations")
    print(f"  3. Non-zero LR coefficients")
    print(f"  4. XGBoost requirements IGNORED (will include XGBoost values when available)")
    print(f"  Total iterations: {total_iterations}")
    
    summary_stats = []
    
    for stance in stances:
        print(f"\nProcessing {stance}...")
        
        # Step 1: Collect all features from all iterations
        all_feature_data = {}  # feature_name -> list of data from each iteration
        
        for i in range(1, total_iterations + 1):
            file_path = os.path.join(base_dir, f"{i:02d}_iteration", f"top_features_{stance}.csv")
            if os.path.exists(file_path):
                df = pd.read_csv(file_path)
                print(f"  Loaded iteration {i:02d}: {len(df)} features")
                
                for _, row in df.iterrows():
                    # Get clean feature name (without *)
                    clean_feature = str(row['feature']).replace('*', '')
                    
                    if clean_feature not in all_feature_data:
                        all_feature_data[clean_feature] = []
                    
                    # Store all data for this feature occurrence
                    feature_occurrence = {
                        'iteration': i,
                        'is_significant': '*' in str(row['feature']),
                        'coefficients': {},
                        'correct_mean_std': row.get('correct_mean_std', ''),
                        'incorrect_mean_std': row.get('incorrect_mean_std', '')
                    }
                    
                    # Store coefficients/importances (only if they are non-zero)
                    for model in models:
                        if (model in df.columns and 
                            pd.notna(row[model]) and 
                            row[model] != '' and 
                            row[model] != 0):
                            feature_occurrence['coefficients'][model] = float(row[model])
                    
                    all_feature_data[clean_feature].append(feature_occurrence)
            else:
                print(f"  Warning: Missing file {file_path}")
        
        print(f"  Total features found: {len(all_feature_data)}")
        
        # Step 2: Apply LR-ONLY filtering
        final_features = []
        
        for feature_name, occurrences in all_feature_data.items():
            # Aggregate data across occurrences
            aggregated_data = {
                'feature_name': feature_name,
                'total_occurrences': len(occurrences),
                'coefficients': {model: [] for model in models},
                'coef_occurrences': {model: 0 for model in models},
                'significances': [],
                'correct_means': [],
                'incorrect_means': [],
                'correct_stds': [],
                'incorrect_stds': []
            }
            
            # Collect data from all occurrences
            for occurrence in occurrences:
                aggregated_data['significances'].append(occurrence['is_significant'])
                
                # Collect coefficients and count valid occurrences
                for model in models:
                    if model in occurrence['coefficients']:
                        aggregated_data['coefficients'][model].append(occurrence['coefficients'][model])
                        aggregated_data['coef_occurrences'][model] += 1
                
                # Parse mean/std values
                if occurrence['correct_mean_std']:
                    correct_mean, correct_std = parse_mean_std(occurrence['correct_mean_std'])
                    if not np.isnan(correct_mean):
                        aggregated_data['correct_means'].append(correct_mean)
                        aggregated_data['correct_stds'].append(correct_std)
                
                if occurrence['incorrect_mean_std']:
                    incorrect_mean, incorrect_std = parse_mean_std(occurrence['incorrect_mean_std'])
                    if not np.isnan(incorrect_mean):
                        aggregated_data['incorrect_means'].append(incorrect_mean)
                        aggregated_data['incorrect_stds'].append(incorrect_std)
            
            # CRITERION 1: Check significance threshold
            num_significant = sum(aggregated_data['significances'])
            if num_significant < min_significance_count:
                print(f"    Skipping {feature_name}: only {num_significant}/{len(occurrences)} significant (need ≥{min_significance_count})")
                continue
            
            # CRITERION 3: Must have non-zero LR coefficients (IGNORE XGBoost requirement)
            lr_coeffs = aggregated_data['coefficients']['Logistic Regression']
            
            if not lr_coeffs:
                print(f"    Skipping {feature_name}: no LR coefficients")
                continue
            
            # CRITERION 2: Check LR coefficient sign consistency
            lr_signs = [1 if x > 0 else -1 if x < 0 else 0 for x in lr_coeffs]
            lr_non_zero_signs = [s for s in lr_signs if s != 0]
            
            if len(set(lr_non_zero_signs)) > 1:
                print(f"    Skipping {feature_name}: LR mixed signs ({len(lr_coeffs)} values)")
                continue
            
            if len(lr_non_zero_signs) == 0:
                print(f"    Skipping {feature_name}: all LR coefficients are zero")
                continue
            
            # LR CRITERIA MET - Calculate final values
            final_lr_coef = np.median(lr_coeffs)
            
            # Include XGBoost values if available (but don't require them)
            xgb_coeffs = aggregated_data['coefficients']['XGBoost']
            final_xgb_importance = np.median(xgb_coeffs) if xgb_coeffs else ''
            
            # Calculate aggregated statistics
            agg_correct_mean = np.mean(aggregated_data['correct_means']) if aggregated_data['correct_means'] else np.nan
            agg_correct_std = np.mean(aggregated_data['correct_stds']) if aggregated_data['correct_stds'] else np.nan
            agg_incorrect_mean = np.mean(aggregated_data['incorrect_means']) if aggregated_data['incorrect_means'] else np.nan
            agg_incorrect_std = np.mean(aggregated_data['incorrect_stds']) if aggregated_data['incorrect_stds'] else np.nan
            
            # Create final entry
            final_features.append({
                'feature': feature_name + '*',  # All included features meet significance threshold
                'Logistic Regression': final_lr_coef,
                'XGBoost': final_xgb_importance,  # Include if available, empty if not
                'correct_mean_std': f"{agg_correct_mean:.4f} ± {agg_correct_std:.4f}" if not np.isnan(agg_correct_mean) else '',
                'incorrect_mean_std': f"{agg_incorrect_mean:.4f} ± {agg_incorrect_std:.4f}" if not np.isnan(agg_incorrect_mean) else '',
                'lr_coef_count': len(lr_coeffs),
                'xgb_coef_count': len(xgb_coeffs),
                'num_significant': num_significant,
                'significance_rate': f"{num_significant}/{len(occurrences)}",
                'lr_sign': 'positive' if lr_non_zero_signs[0] > 0 else 'negative',
                'has_xgb': len(xgb_coeffs) > 0
            })
        
        # Step 3: Sort by significance count, then by LR coefficient count
        final_features.sort(key=lambda x: (x['num_significant'], x['lr_coef_count']), reverse=True)
        
        # Remove helper columns for final output
        final_features_clean = []
        for feature in final_features:
            clean_feature = {k: v for k, v in feature.items() 
                           if k not in ['lr_coef_count', 'xgb_coef_count', 'num_significant', 'significance_rate', 'lr_sign', 'has_xgb']}
            final_features_clean.append(clean_feature)
        
        # Step 4: Save final aggregated file
        final_df = pd.DataFrame(final_features_clean)
        output_file = os.path.join(output_dir, f"final_top_features_{stance}.csv")
        final_df.to_csv(output_file, index=False)
        
        # Step 5: Save detailed info file with occurrence counts
        detailed_df = pd.DataFrame(final_features)
        detailed_file = os.path.join(output_dir, f"detailed_features_{stance}.csv")
        detailed_df.to_csv(detailed_file, index=False)
        
        # Collect summary statistics
        total_features = len(all_feature_data)
        final_count = len(final_features)
        positive_lr = len([f for f in final_features if f['lr_sign'] == 'positive'])
        negative_lr = len([f for f in final_features if f['lr_sign'] == 'negative'])
        with_xgb = len([f for f in final_features if f['has_xgb']])
        
        summary_stats.append({
            'Stance': stance,
            'Total_Features': total_features,
            'Final_Features': final_count,
            'Positive_LR_Coef': positive_lr,
            'Negative_LR_Coef': negative_lr,
            'Also_Have_XGB': with_xgb,
            'LR_Only': final_count - with_xgb,
            'Avg_Significance': round(np.mean([f['num_significant'] for f in final_features]) if final_features else 0, 1),
            'Success_Rate': f"{final_count}/{total_features}",
            'Success_Percentage': f"{100*final_count/total_features:.1f}%" if total_features > 0 else "0%"
        })
        
        print(f"  Final features meeting LR criteria: {final_count}")
        print(f"  Positive LR coefficients: {positive_lr}")
        print(f"  Negative LR coefficients: {negative_lr}")
        print(f"  Also have XGBoost values: {with_xgb}")
        print(f"  LR-only features: {final_count - with_xgb}")
        print(f"  Average significance: {round(np.mean([f['num_significant'] for f in final_features]) if final_features else 0, 1)}/{total_iterations}")
        print(f"  Success rate: {100*final_count/total_features:.1f}%")
        print(f"  Saved to: {output_file}")
        print(f"  Detailed info saved to: {detailed_file}")
    
    # Save summary statistics
    summary_df = pd.DataFrame(summary_stats)
    summary_file = os.path.join(output_dir, f"aggregation_summary_lr_only_{min_significance_count}sig.csv")
    summary_df.to_csv(summary_file, index=False)
    
    print(f"\n" + "="*80)
    print(f"AGGREGATION SUMMARY - LR-ONLY FILTERING")
    print("="*80)
    print(summary_df.to_string(index=False))
    print(f"\nSummary saved to: {summary_file}")
    
    return summary_df

# Main execution with configurable parameters
if __name__ == '__main__':
    base_dir = ""
    output_dir = "important_features"
    
    # CONFIGURABLE PARAMETERS - CHANGE THESE AS NEEDED
    MIN_SIGNIFICANCE_COUNT = 7   # Feature must be significant in at least 7 iterations  
    TOTAL_ITERATIONS = 13        # Total number of iterations
    
    print("Creating final aggregated feature files with LR-ONLY filtering...")
    print("LR-based criteria (XGBoost ignored):")
    print("1. Statistically significantly different in ≥7 iterations")
    print("2. Same LR coefficient sign across iterations") 
    print("3. Non-zero LR coefficients")
    print("4. XGBoost values included when available but not required")
    
    summary = create_final_aggregated_features_lr_only(
        base_dir=base_dir,
        output_dir=output_dir,
        min_significance_count=MIN_SIGNIFICANCE_COUNT,
        total_iterations=TOTAL_ITERATIONS
    )
    
    print(f"\n✅ LR-only filtering completed!")
    print(f"📁 Created files:")
    print(f"   - final_top_features_*.csv (4 files) - features meeting LR criteria")
    print(f"   - detailed_features_*.csv (4 files with detailed diagnostics)")
    print(f"   - aggregation_summary_lr_only_{MIN_SIGNIFICANCE_COUNT}sig.csv")
```

# Two Filter


1. Same LR coefficient sign across above ≥7 iterations AND
2. Exist in BOTH LR and XGBoost with non-zero values

```python
import pandas as pd
import numpy as np
import os
import warnings
warnings.filterwarnings('ignore')

def parse_mean_std(mean_std_str):
    """Parse '0.1234 ± 0.0567' format"""
    if pd.isna(mean_std_str) or mean_std_str == '':
        return np.nan, np.nan
    parts = str(mean_std_str).split(' ± ')
    if len(parts) == 2:
        return float(parts[0]), float(parts[1])
    return np.nan, np.nan

def create_final_aggregated_features_occurrence_based(base_dir, output_dir, 
                                                    min_occurrence_count=7,
                                                    total_iterations=13):
    """
    Aggregate features based on occurrence frequency (not statistical significance):
    1. Exist in BOTH LR and XGBoost in ≥MIN_OCCURRENCE_COUNT iterations
    2. Same LR coefficient sign across those iterations
    3. Statistical significance requirement REMOVED
    
    Parameters:
    -----------
    base_dir : str
        Base directory containing iteration folders
    output_dir : str  
        Output directory for final files
    min_occurrence_count : int
        Minimum number of iterations where feature must exist in both models (default: 7)
    total_iterations : int
        Total number of iterations (default: 13)
    """
    stances = ["All_Stances", "FAVOR", "AGAINST", "NONE"]
    models = ["Logistic Regression", "XGBoost"]
    
    os.makedirs(output_dir, exist_ok=True)
    
    print(f"Configuration - OCCURRENCE-BASED FILTERING:")
    print(f"  1. Exist in BOTH LR and XGBoost in ≥{min_occurrence_count} iterations")
    print(f"  2. Same LR coefficient sign across those iterations")
    print(f"  3. Statistical significance requirement REMOVED")
    print(f"  Total iterations: {total_iterations}")
    
    summary_stats = []
    
    for stance in stances:
        print(f"\nProcessing {stance}...")
        
        # Step 1: Collect all features from all iterations
        all_feature_data = {}  # feature_name -> list of data from each iteration
        
        for i in range(1, total_iterations + 1):
            file_path = os.path.join(base_dir, f"{i:02d}_iteration", f"top_features_{stance}.csv")
            if os.path.exists(file_path):
                df = pd.read_csv(file_path)
                print(f"  Loaded iteration {i:02d}: {len(df)} features")
                
                for _, row in df.iterrows():
                    # Get clean feature name (without *)
                    clean_feature = str(row['feature']).replace('*', '')
                    
                    if clean_feature not in all_feature_data:
                        all_feature_data[clean_feature] = []
                    
                    # Store all data for this feature occurrence
                    feature_occurrence = {
                        'iteration': i,
                        'is_significant': '*' in str(row['feature']),
                        'coefficients': {},
                        'correct_mean_std': row.get('correct_mean_std', ''),
                        'incorrect_mean_std': row.get('incorrect_mean_std', '')
                    }
                    
                    # Store coefficients/importances (only if they are non-zero)
                    for model in models:
                        if (model in df.columns and 
                            pd.notna(row[model]) and 
                            row[model] != '' and 
                            row[model] != 0):
                            feature_occurrence['coefficients'][model] = float(row[model])
                    
                    all_feature_data[clean_feature].append(feature_occurrence)
            else:
                print(f"  Warning: Missing file {file_path}")
        
        print(f"  Total features found: {len(all_feature_data)}")
        
        # Step 2: Apply occurrence-based filtering
        final_features = []
        
        for feature_name, occurrences in all_feature_data.items():
            # Aggregate data across occurrences
            aggregated_data = {
                'feature_name': feature_name,
                'total_occurrences': len(occurrences),
                'coefficients': {model: [] for model in models},
                'coef_occurrences': {model: 0 for model in models},
                'both_models_occurrences': 0,  # Count iterations where both models have values
                'significances': [],
                'correct_means': [],
                'incorrect_means': [],
                'correct_stds': [],
                'incorrect_stds': []
            }
            
            # Collect data from all occurrences and count "both models" occurrences
            for occurrence in occurrences:
                aggregated_data['significances'].append(occurrence['is_significant'])
                
                # Check if both models have coefficients in this iteration
                has_lr = 'Logistic Regression' in occurrence['coefficients']
                has_xgb = 'XGBoost' in occurrence['coefficients']
                
                if has_lr and has_xgb:
                    aggregated_data['both_models_occurrences'] += 1
                
                # Collect coefficients and count valid occurrences
                for model in models:
                    if model in occurrence['coefficients']:
                        aggregated_data['coefficients'][model].append(occurrence['coefficients'][model])
                        aggregated_data['coef_occurrences'][model] += 1
                
                # Parse mean/std values
                if occurrence['correct_mean_std']:
                    correct_mean, correct_std = parse_mean_std(occurrence['correct_mean_std'])
                    if not np.isnan(correct_mean):
                        aggregated_data['correct_means'].append(correct_mean)
                        aggregated_data['correct_stds'].append(correct_std)
                
                if occurrence['incorrect_mean_std']:
                    incorrect_mean, incorrect_std = parse_mean_std(occurrence['incorrect_mean_std'])
                    if not np.isnan(incorrect_mean):
                        aggregated_data['incorrect_means'].append(incorrect_mean)
                        aggregated_data['incorrect_stds'].append(incorrect_std)
            
            # FILTER 1: Must exist in both models in at least MIN_OCCURRENCE_COUNT iterations
            if aggregated_data['both_models_occurrences'] < min_occurrence_count:
                print(f"    Skipping {feature_name}: exists in both models in only {aggregated_data['both_models_occurrences']}/{len(occurrences)} iterations (need ≥{min_occurrence_count})")
                continue
            
            # FILTER 2: Must have LR coefficients in at least MIN_OCCURRENCE_COUNT iterations
            lr_coeffs = aggregated_data['coefficients']['Logistic Regression']
            if len(lr_coeffs) < min_occurrence_count:
                print(f"    Skipping {feature_name}: LR coefficients in only {len(lr_coeffs)} iterations (need ≥{min_occurrence_count})")
                continue
            
            # FILTER 3: Must have XGBoost importance in at least MIN_OCCURRENCE_COUNT iterations  
            xgb_coeffs = aggregated_data['coefficients']['XGBoost']
            if len(xgb_coeffs) < min_occurrence_count:
                print(f"    Skipping {feature_name}: XGBoost importance in only {len(xgb_coeffs)} iterations (need ≥{min_occurrence_count})")
                continue
            
            # FILTER 4: Check LR coefficient sign consistency across MIN_OCCURRENCE_COUNT iterations
            lr_signs = [1 if x > 0 else -1 if x < 0 else 0 for x in lr_coeffs]
            lr_non_zero_signs = [s for s in lr_signs if s != 0]
            
            if len(set(lr_non_zero_signs)) > 1:
                print(f"    Skipping {feature_name}: LR mixed signs across {len(lr_coeffs)} iterations")
                continue
            
            if len(lr_non_zero_signs) == 0:
                print(f"    Skipping {feature_name}: all LR coefficients are zero")
                continue
            
            # ALL CRITERIA MET - Calculate final values
            final_lr_coef = np.median(lr_coeffs)
            final_xgb_importance = np.median(xgb_coeffs)
            
            # Calculate aggregated statistics
            agg_correct_mean = np.mean(aggregated_data['correct_means']) if aggregated_data['correct_means'] else np.nan
            agg_correct_std = np.mean(aggregated_data['correct_stds']) if aggregated_data['correct_stds'] else np.nan
            agg_incorrect_mean = np.mean(aggregated_data['incorrect_means']) if aggregated_data['incorrect_means'] else np.nan
            agg_incorrect_std = np.mean(aggregated_data['incorrect_stds']) if aggregated_data['incorrect_stds'] else np.nan
            
            # Count significance occurrences (for info only, not filtering)
            num_significant = sum(aggregated_data['significances'])
            
            # Create final entry
            final_features.append({
                'feature': feature_name,  # No * since significance not used for filtering
                'Logistic Regression': final_lr_coef,
                'XGBoost': final_xgb_importance,
                'correct_mean_std': f"{agg_correct_mean:.4f} ± {agg_correct_std:.4f}" if not np.isnan(agg_correct_mean) else '',
                'incorrect_mean_std': f"{agg_incorrect_mean:.4f} ± {agg_incorrect_std:.4f}" if not np.isnan(agg_incorrect_mean) else '',
                'lr_coef_count': len(lr_coeffs),
                'xgb_coef_count': len(xgb_coeffs),
                'both_models_count': aggregated_data['both_models_occurrences'],
                'num_significant': num_significant,
                'significance_rate': f"{num_significant}/{len(occurrences)}",
                'lr_sign': 'positive' if lr_non_zero_signs[0] > 0 else 'negative'
            })
        
        # Step 3: Sort by both_models_count, then by individual model counts
        final_features.sort(key=lambda x: (x['both_models_count'], x['lr_coef_count'], x['xgb_coef_count']), reverse=True)
        
        # Remove helper columns for final output
        final_features_clean = []
        for feature in final_features:
            clean_feature = {k: v for k, v in feature.items() 
                           if k not in ['lr_coef_count', 'xgb_coef_count', 'both_models_count', 'num_significant', 'significance_rate', 'lr_sign']}
            final_features_clean.append(clean_feature)
        
        # Step 4: Save final aggregated file
        final_df = pd.DataFrame(final_features_clean)
        output_file = os.path.join(output_dir, f"final_top_features_{stance}.csv")
        final_df.to_csv(output_file, index=False)
        
        # Step 5: Save detailed info file with occurrence counts
        detailed_df = pd.DataFrame(final_features)
        detailed_file = os.path.join(output_dir, f"detailed_features_{stance}.csv")
        detailed_df.to_csv(detailed_file, index=False)
        
        # Collect summary statistics
        total_features = len(all_feature_data)
        final_count = len(final_features)
        positive_lr = len([f for f in final_features if f['lr_sign'] == 'positive'])
        negative_lr = len([f for f in final_features if f['lr_sign'] == 'negative'])
        avg_both_models = round(np.mean([f['both_models_count'] for f in final_features]) if final_features else 0, 1)
        
        summary_stats.append({
            'Stance': stance,
            'Total_Features': total_features,
            'Final_Features': final_count,
            'Positive_LR_Coef': positive_lr,
            'Negative_LR_Coef': negative_lr,
            'Avg_Both_Models_Count': avg_both_models,
            'Success_Rate': f"{final_count}/{total_features}",
            'Success_Percentage': f"{100*final_count/total_features:.1f}%" if total_features > 0 else "0%"
        })
        
        print(f"  Final features meeting criteria: {final_count}")
        print(f"  Positive LR coefficients: {positive_lr}")
        print(f"  Negative LR coefficients: {negative_lr}")
        print(f"  Average both-models occurrences: {avg_both_models}/{total_iterations}")
        print(f"  Success rate: {100*final_count/total_features:.1f}%")
        print(f"  Saved to: {output_file}")
        print(f"  Detailed info saved to: {detailed_file}")
    
    # Save summary statistics
    summary_df = pd.DataFrame(summary_stats)
    summary_file = os.path.join(output_dir, f"aggregation_summary_occurrence_{min_occurrence_count}.csv")
    summary_df.to_csv(summary_file, index=False)
    
    print(f"\n" + "="*80)
    print(f"AGGREGATION SUMMARY - OCCURRENCE-BASED FILTERING (≥{min_occurrence_count} iterations)")
    print("="*80)
    print(summary_df.to_string(index=False))
    print(f"\nSummary saved to: {summary_file}")
    
    return summary_df

# Main execution with configurable parameters
if __name__ == '__main__':
    base_dir = ""
    output_dir = "important_features"
    
    # CONFIGURABLE PARAMETERS - CHANGE THESE AS NEEDED
    MIN_OCCURRENCE_COUNT = 1     # Feature must exist in both models in at least 7 iterations
    TOTAL_ITERATIONS = 13        # Total number of iterations
    
    print("Creating final aggregated feature files with OCCURRENCE-BASED filtering...")
    print("Criteria:")
    print(f"1. Exist in BOTH LR and XGBoost in ≥{MIN_OCCURRENCE_COUNT} iterations")
    print("2. Same LR coefficient sign across those iterations") 
    print("3. Statistical significance requirement REMOVED")
    
    summary = create_final_aggregated_features_occurrence_based(
        base_dir=base_dir,
        output_dir=output_dir,
        min_occurrence_count=MIN_OCCURRENCE_COUNT,
        total_iterations=TOTAL_ITERATIONS
    )
    
    print(f"\n✅ Occurrence-based filtering completed!")
    print(f"📁 Created files:")
    print(f"   - final_top_features_*.csv (4 files) - features existing in both models ≥{MIN_OCCURRENCE_COUNT} times")
    print(f"   - detailed_features_*.csv (4 files with detailed diagnostics)")
    print(f"   - aggregation_summary_occurrence_{MIN_OCCURRENCE_COUNT}.csv")
```
