```python
import pandas as pd
df = pd.read_csv("all_combined.csv")
```

```python
df_wtwt = df[df['dataset'] == 'wtwt']
```

```python

stance_mapping = {
    'unrelated': 'NONE',
    'comment': 'NONE',
    'support': 'FAVOR',
    'refute': 'AGAINST'
}
df_wtwt['stance'] = df_wtwt['stance'].replace(stance_mapping)
results = {}
for type_val in ['train', 'test']:
    df_filtered = df_wtwt[df_wtwt['type'] == type_val]
    total_count = len(df_filtered)

    if total_count > 0:
        stance_counts = df_filtered['stance'].value_counts()
        stance_percentages = (stance_counts / total_count) * 100

        results[type_val] = {
            'counts': stance_counts.to_dict(),
            'percentages': stance_percentages.to_dict()
        }
    else:
        results[type_val] = {
            'counts': {},
            'percentages': {}
        }

# Print the results
for type_val, data in results.items():
    print(f"Type: {type_val}")
    print("Counts:")
    for stance, count in data['counts'].items():
        print(f"  {stance}: {count}")
    print("Percentages:")
    for stance, percentage in data['percentages'].items():
        print(f"  {stance}: {percentage:.2f}%")
    print("-" * 30)
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("all_combined.csv")

# Map the 'stance' column for the entire dataset
stance_mapping = {
    'unrelated': 'NONE',
    'comment': 'NONE',
    'support': 'FAVOR',
    'refute': 'AGAINST'
}
df['stance'] = df['stance'].replace(stance_mapping)

# Process for 'train' type
df_train = df[df['type'] == 'train']
train_stance_counts = df_train['stance'].value_counts()
train_total_points = len(df_train)
train_stance_percentages = (train_stance_counts / train_total_points) * 100

print("--- Stance Distribution for 'train' data ---")
print("Counts:")
print(train_stance_counts)
print("\nPercentages:")
print(train_stance_percentages)
print("-" * 40)

# Process for 'test' type
df_test = df[df['type'] == 'test']
test_stance_counts = df_test['stance'].value_counts()
test_total_points = len(df_test)
test_stance_percentages = (test_stance_counts / test_total_points) * 100

print("\n--- Stance Distribution for 'test' data ---")
print("Counts:")
print(test_stance_counts)
print("\nPercentages:")
print(test_stance_percentages)
print("-" * 40)
```
