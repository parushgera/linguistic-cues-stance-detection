```python
mpqa_lexicon = {"certainty": set(), "doubt": set(), "hedging": set()}

with open("subjclueslen1-HLTEMNLP05.tff", "r") as f:
    for line in f:
        parts = line.strip().split()
        
        # Extract word
        word = None
        strength = None
        polarity = None
        
        for part in parts:
            if part.startswith("word1="):
                word = part.split("=")[1]
            elif part.startswith("strength="):
                strength = part.split("=")[1]
            elif part.startswith("priorpolarity="):
                polarity = part.split("=")[1]

        # If any required value is missing, skip the line
        if not word or not strength or not polarity:
            continue

        # Categorize words into lexicon
        if strength == "strongsubj" and polarity in ["positive", "neutral"]:
            mpqa_lexicon["certainty"].add(word)
        elif strength == "weaksubj" and polarity == "neutral":
            mpqa_lexicon["hedging"].add(word)
        elif polarity == "negative":
            mpqa_lexicon["doubt"].add(word)

print("✅ Loaded MPQA Certainty Words:", len(mpqa_lexicon["certainty"]))
print("✅ Loaded MPQA Doubt Words:", len(mpqa_lexicon["doubt"]))
print("✅ Loaded MPQA Hedging Words:", len(mpqa_lexicon["hedging"]))
```

```python
# Load NRC Emotion Lexicon
nrc_emotion_lexicon = {"anger": set(), "joy": set(), "fear": set(), "sadness": set()}

with open("NRC-Emotion-Lexicon-Wordlevel-v0.92.txt", "r") as f:
    for line in f:
        parts = line.strip().split("\t")
        if len(parts) == 3:
            word, emotion, score = parts
            if score == "1":  # Only include words with emotion score of 1
                if emotion in nrc_emotion_lexicon:
                    nrc_emotion_lexicon[emotion].add(word)

print("Loaded NRC Emotion Words:", {k: len(v) for k, v in nrc_emotion_lexicon.items()})
```
